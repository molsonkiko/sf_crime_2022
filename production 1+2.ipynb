```python
unit_per_year = {
    '': float('nan'),
    'Hour': 24 * 365.,
    'Day': 365.,
    'Week': 52.,
    'Month': 12.,
    'Year': 1.,
    'Y+1': 366/365,
}

def term_to_float(term, from_unit, to_unit = 'Day'):
    return try_float(term) * unit_per_year[to_unit] / unit_per_year[from_unit] 
```

```python
unit_vars = list(prod1on2.filter(regex='(?i)unit\s*$', axis=1).columns)
term_vars = []
for uvar in unit_vars:
    term_vars.append(list(prod1on2.filter(regex=re.sub('(?i)\s*unit', '', uvar), axis=1).columns)[0])
    print(uvar, term_vars[-1])
    print(prod1on2[uvar].unique())
    prod1on2[term_vars[-1]] = prod1on2.apply(lambda x: term_to_float(x[term_vars[-1]], 
                                                            x[uvar].strip()), axis=1)
prod1on2.drop(unit_vars, inplace=True, axis=1, errors='ignore')
```

```python
def change_infoframe_terminology(info):
    new_info = info.rename({'nanpct': 'Fraction missing values',
                           'nunique': 'num unique values'},
                          axis=1)
    new_info['dtype'] = new_info['dtype'].apply(str).map({'int64': 'Whole Number',
                                                          'float64': 'Decimal number',
                                                          'object': 'text',
                                                          'datetime64[ns]': 'Date + time'})
    return new_info
```

```python
all_join_description = change_infoframe_terminology(make_infoframe(all_join))
all_join_description.to_csv('output files/'
                            'SUMMARY STATISTICS incident reports JOIN da caseload JOIN prod1 JOIN prod2.csv')
```
