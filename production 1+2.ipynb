```python
unit_per_year = {
    '': float('nan'),
    'Hour': 24 * 365.,
    'Day': 365.,
    'Week': 52.,
    'Month': 12.,
    'Year': 1.,
    'Y+1': 366/365,
}

def term_to_float(term, from_unit, to_unit = 'Day'):
    return try_float(term) * unit_per_year[to_unit] / unit_per_year[from_unit] 
```

```python
unit_vars = ['SP Term Unit', 'CJ Term Unit', 'SupervisionTerm Unit', 'Program Term Unit']
term_vars = ['SP Term', 'CJ Term', 'Supervision Term', 'Program Term']
for uvar, var in zip(unit_vars, term_vars):
    prod1on2[term_vars[-1]] = prod1on2.apply(lambda x: term_to_float(x[term_vars[-1]], 
                                                            x[uvar].strip()), axis=1)
prod1on2.drop(unit_vars, inplace=True, axis=1, errors='ignore')
```

```python
def zero_if_dismissed(x, dismissed):
    if dismissed:
        return 0
    return x

prod1on2['dismissed'] = prod1on2['Outcome Sub-Type Description'].str.contains('(?i)dism', regex=True) \
                        | prod1on2['Outcome Description'].str.contains('(?i)dism', regex=True)

for var in term_vars:
    prod1on2[var] = prod1on2.apply(lambda x: zero_if_dismissed(x[var], x['dismissed']), axis=1)
```

```python
def change_infoframe_terminology(info):
    new_info = info.rename({'nanpct': 'Fraction missing values',
                           'nunique': 'num unique values'},
                          axis=1)
    new_info['dtype'] = new_info['dtype'].apply(str).map({'int64': 'Whole Number',
                                                          'float64': 'Decimal number',
                                                          'object': 'text',
                                                          'datetime64[ns]': 'Date + time'})
    return new_info
```

1.   For all cases with disposition dates of January 1, 2020 until the present, let’s extract those cases that have the following incident codes:  3011, 3012, 3013, 3014, 3411, 3412, 3413, 3414, which are categorized as street robberies, and attempted street robberies.

2.   For this subset of street robbery offenses, let’s put all the dismissed or discharged cases at the top of our list.  Would be helpful to see how many cases were dumped by Chesa.

3.   Following the discharged cases, let’s have all the cases that have a specific county jail sentences assigned, where they tell us the actual number of days the defendant got

4.   Next on the list are those with specific state prison sentences, where they provide the actual length of the sentence

5.   Next are those that don’t have a specific county jail term, but it’s indicated that the disposition is “county jail” or “county jail with probation.”

5.   Next are those that don’t have a specific state prison sentence, but it’s indicated that the disposition is “state prison.”

6.   Last on the list are those street robbery cases with any other types of dispositions like “diversion.”

```python
all_join['Year'] = all_join['Incident Date'].str.slice(0, 4).astype(int)
```

```python
robberies = all_join[all_join['Incident Code'].isin({3011, 3012, 3013,
                                                    3014, 3411, 3412, 3413, 3414})] \
                    .reset_index(drop=True)
```

```python
robberies_by_dism_yr = robberies.groupby(['Year', 'dismissed']).index.count()
robberies_by_dism_yr.to_excel('output files/' +
                            'Number of robberies by year and dismissal status.xlsx')
robberies_by_dism_yr
```

```python
def make_rank_dism_jail(row):
    if row['dismissed']:
        return 0
    try:
        if not (np.isnan(row['CJ Term']) or row['CJ Term'] == 0):
            return 1
    except TypeError:
#         print(repr(row['CJ Term']))
        pass
    try:
        if not (np.isnan(row['SP Term']) or row['SP Term'] == 0):
            return 2
    except TypeError:
#         print(repr(row['SP Term']))
        pass
    if re.search('county\s*jail', row['Dispo Description'], re.IGNORECASE):
        return 3
    if re.search('state\s*prison', row['Dispo Description'], re.IGNORECASE):
        return 4
    return 5
```

```python
def split_by_years(df, excel_fname):
    new_df = df.copy()
    new_df['ordering'] = new_df.apply(make_rank_dism_jail, axis=1)
    df_2018_19 = new_df[new_df.Year.isin({2018, 2019})].reset_index(drop=True)
    df_2020_22 = new_df[new_df.Year > 2019].reset_index(drop=True)
    uniques = pd.Series(new_df['Incident Category'].unique()).sort_values()
    with pd.ExcelWriter(excel_fname, 
                        engine="openpyxl", 
                        mode="a", 
                        if_sheet_exists="replace") as writer:  
        df_2018_19.to_excel(writer, sheet_name='2018-2019', index=False)
        df_2020_22.to_excel(writer, sheet_name='2020-2021', index=False)
        uniques.to_excel(writer, sheet_name='all unique incident categories', 
                         index=False,
                         header = ['Incident Category'])
    return df_2018_19, df_2020_22
```

```python
all_2018, all_2020 = split_by_years(all_join, 'output files/incident reports JOIN da caseload'
                                    ' JOIN prod1 JOIN prod2 BY YEAR.xlsx')
```

```python
rob_2018, rob_2020 = split_by_years(robberies, 'output files/robberies BY YEAR.xlsx')
```

```python
robberies['ordering'] = robberies.apply(make_rank_dism_jail, axis=1)
robberies.sort_values('ordering').to_excel('output files/' +
                                         'Robberies ordered by dismissal and dispo.xlsx',
                                        index=False)
```
